**Questão 2**. Pythonaldo e Javanildo, dois alunos do BTI, estavam discutindo sobre suas popularidades nas redes sociais. Pythonaldo possui 231 contatos no Facebook enquanto que Javanildo possui 180 contatos no Tinder. Sabe-se que:

- o número de contatos no Facebook e Tinder segue uma distribuição normal. 
- a média e o desvio padrão para o número de contatos no Facebook são respectivamente, 250 e 36.
- a média e o desvio padrão para o número de contatos no Tinder são respectivamente, 200 e 80.
 
a) (**3 pontos**) Utilizando o conceito de Z-Score, quem seria mais popular, Pythonaldo ou Javanildo?


```python
import scipy.stats as stats

pythonaldo = 231
media_fb = 250
dp_fb = 36

javanildo = 180
media_tin = 200
dp_tin = 80

cdf_p = stats.norm.cdf(pythonaldo, media_fb, dp_fb)
cdf_j = stats.norm.cdf(javanildo, media_tin, dp_tin)

print('Pythonaldo: ', cdf_p,'\nJavanildo: ', cdf_j)
```

O mais popular é o que possui um maior z-score em relação a média. Logo, Javanildo é o mais popular.
