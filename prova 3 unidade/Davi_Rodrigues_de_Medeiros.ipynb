```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dados = pd.read_csv('CopaMundoGeral.csv', sep=';')
dados.head()
```

a) (1,5 pontos) Encontre a assimetria da distribuição de Torcedores e justifique/interprete o resultado a partir de um gráfico. (sugestão: utilizar a função sns.distplot). O texto da justificativa/interpretação deverá ser realizado em formato de Markdown.

```python
sns.distplot(dados['Torcedores'], bins=20)
```

No Gráfico acima, podemos observar que a distribuição de torcedores continua 'constante' durante esse tempo de jogos. Há pequenos picos, mas a média varia muito pouco.

Abaixo, podemos analisar mais a fundo o número de torcedores durante todo esses anos.

```python
dados = dados.sort_values('Ano')
sns.set(font_scale=1.3)

plt.bar(list(dados['Ano'].unique()), 
        list(dados['Torcedores']))
plt.title('Quantidade de torcedores por ano')
```

No gráfico acima, é perceptível que o número de trocedores cresceu ao logo dos anos.

b) (1,5 pontos) A partir da análise do curtose da distribuição de Gols pode-se chegar a qual conclusão? Justifique a sua resposta baseada em um gráfico de sua escolha e no valor do curtose. O texto da justificativa deverá ser realizado em formato de Markdown.

```python
plt.plot(x=dados['Ano'], y=[dados['Gols'], dados['Torcedores']])
```

c) (1,0 ponto) Utilizando como base um gráfico de caixa (box-plot) faça uma análise sobre os valores da média e mediana das distribuições de Gols e Torcedores. Adicionalmente, utilize as informações da variância e desvio padrão dessas distribuições para reforçar as suas conclusões. O texto da análise deverá ser realizado em formato de Markdown.

```python
fig, ax = plt.subplots(figsize=(10,5))

plt.suptitle("Média (vermelho) vs Mediana (verde)", fontsize=20, y = 1.09)
ax.set_title("Gols", fontsize=12)
ax.axvline(dados['Gols'].mean(), color="r")
ax.axvline(dados['Gols'].median(), color="g")
sns.distplot(dados['Gols'], kde = False,bins=20)
plt.show()
```

No gráfico acima (Número de Gols) podemos observar que a média e a mediana se encontram bem próximas. Isso é justificado de acordo com a distribuição dos gols, que possui gols concentrados entre 1.0 e 2.0

```python
fig, ax = plt.subplots(figsize=(10,5))
plt.suptitle("Média (vermelho) vs Mediana (verde)", fontsize=15, y = 1.09)
ax.set_title("Torcedores", fontsize=12)
ax.axvline(dados['Torcedores'].mean(), color="r")
ax.axvline(dados['Torcedores'].median(), color="g")
sns.distplot(dados['Torcedores'], kde = False,bins=20)
plt.show()
```

No gráfico acima (Número de Torcedores) já é observado que há uma pequena diferença de distancia entre a média e mediana, isso se da por que existe um valor como mostrado acima, que está em 3.0, ou seja, fora do ponto de concentração.
